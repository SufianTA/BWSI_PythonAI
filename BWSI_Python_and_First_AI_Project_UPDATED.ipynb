## Part 3. Variables

A **variable** is a named place to store information.

You can think of a variable like a labeled box.

If we write:

`score = 95`

that means Python should store the value `95` inside a box called `score`.

````python

# Create a variable called name and store text inside it.
name = "Sofia"

# Create a variable called grade and store a number inside it.
grade = 10

# print() can show both variables.
print(name)
print(grade)

# Python looks up the values stored in the variables and shows them.
````

### Strings, integers, and floats

Common data types you will use often:

- **string**: text, such as `"hello"`
- **integer**: whole number, such as `7`
- **float**: decimal number, such as `3.14`

````python

# A string stores text.
school = "BWSI"

# An integer stores a whole number.
year = 2026

# A float stores a decimal number.
temperature = 72.5

print(school)
print(year)
print(temperature)
````

````python

# You can also combine text and variables in a readable way with f-strings.
# The letter f before the quotation marks lets us place variables inside braces.

student_name = "Liam"
favorite_number = 8

print(f"My name is {student_name}.")
print(f"My favorite number is {favorite_number}.")
````

## Part 4. Basic math in Python

Python can do arithmetic just like a calculator.

````python

# Store two numbers in variables.
a = 12
b = 5

# Add the numbers.
print(a + b)

# Subtract the numbers.
print(a - b)

# Multiply the numbers.
print(a * b)

# Divide the numbers.
print(a / b)

# Exponents use two stars.
print(a ** 2)
````

````python

# You can save the result of a calculation in a new variable.
length = 7
width = 4

area = length * width

print(f"The area is {area}.")
````

## Part 5. Getting input from the user

Programs become more interesting when they respond to a user.

`input()` lets the user type something.

````python

# input() waits for the user to type a response.
# Whatever the user types is stored as text.

user_name = input("Type your name: ")

# We then use that value in a message.
print(f"Nice to meet you, {user_name}!")
````

### Important detail about `input()`

`input()` gives back a **string**.

That means even if a user types `12`, Python sees it as text first, not as a number.

If you want a number, you usually convert it with `int()` or `float()`.

````python

# Ask for the user's age.
age_text = input("How old are you? ")

# Convert the text into an integer.
age_number = int(age_text)

# Use the number in a calculation.
age_next_year = age_number + 1

print(f"Next year you will be {age_next_year}.")
````

## Part 6. Making decisions with if statements

A program can choose what to do based on a condition.

That is where `if` statements come in.

````python

# Store a number in a variable.
temperature = 85

# The if statement checks whether the condition is true.
if temperature > 80:
    print("It is hot outside.")

# Because 85 is greater than 80, Python prints the message.
````

### Why indentation matters

Python uses indentation to show which lines belong together.

The indented line under `if` only runs if the condition is true.

In Python, spacing is not decoration.  
Spacing is part of the language.

````python

temperature = 65

if temperature > 80:
    print("It is hot outside.")
else:
    print("It is not hot outside.")
````

````python

# We can also check several possibilities.

score = 91

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: D or below")
````

## Part 7. Repeating actions with loops

A **loop** repeats code.

One common loop is a `for` loop.

````python

# range(5) gives the numbers 0, 1, 2, 3, and 4.
for number in range(5):
    print(number)
````

### Reading that loop in plain English

`for number in range(5):`

means

repeat the next indented block one time for each number from 0 to 4.

````python

# This loop prints a message five times.
for number in range(5):
    print(f"Loop number {number}")
````

````python

# We can start and stop at specific numbers.
for number in range(1, 6):
    print(number)
````

## Part 8. Lists

A **list** stores many items in one place.

Lists are useful when you want to keep a group of related values together.

````python

# Create a list of favorite subjects.
subjects = ["math", "science", "coding", "art"]

# Print the whole list.
print(subjects)

# Print the first item in the list.
# Python counting starts at 0, not 1.
print(subjects[0])

# Print the third item in the list.
print(subjects[2])
````

````python

# We can loop through a list one item at a time.
subjects = ["math", "science", "coding", "art"]

for subject in subjects:
    print(f"I can study {subject}.")
````

````python

# We can add a new item to a list with append().
subjects = ["math", "science", "coding"]

subjects.append("history")

print(subjects)
````

## Part 9. Functions

Functions help us organize code.

They let us give a name to a useful set of steps.

That means we can write the steps once and use them many times.

````python

# Define a function called greet_student.
def greet_student(name):
    # This line runs when the function is called.
    print(f"Hello, {name}! Welcome to Python.")

# Call the function with one value.
greet_student("Ava")

# Call the same function again with a different value.
greet_student("Noah")
````

### What is happening here

`def greet_student(name):`

means

create a function named `greet_student` that expects one input called `name`.

Inside the function, we use that input to build a message.

````python

# Functions can also return values.
def square_number(number):
    # Multiply the number by itself.
    answer = number * number

    # return sends the result back to the place where the function was called.
    return answer

result = square_number(6)

print(result)
````

## Part 10. Dictionaries

A **dictionary** stores pairs of information.

Each pair has:

- a **key**
- a **value**

This is useful when you want to match one thing to another.

````python

# Create a dictionary that maps subjects to room numbers.
rooms = {
    "math": 101,
    "science": 202,
    "art": 305
}

# Look up the value connected to a key.
print(rooms["science"])
````

````python

# Loop through each key and value in a dictionary.
rooms = {
    "math": 101,
    "science": 202,
    "art": 305
}

for subject, room_number in rooms.items():
    print(f"{subject} is in room {room_number}.")
````

## Part 12. What is AI?

Artificial Intelligence, or AI, is when we teach a computer how to make a decision based on examples.

A normal computer program follows exact instructions written by a person.

For example:

```python
if temperature > 90:
    print("It is hot")
```

That is not really AI.
That is just a rule that a human wrote.

AI is different.

With AI, instead of writing every rule by hand, we give the computer examples and let it learn patterns.

For example, imagine you show a computer many text messages:

- "I am so excited today" → happy
- "This is the worst day ever" → not happy
- "I love this" → happy
- "I feel tired and upset" → not happy

After seeing enough examples, the computer starts noticing patterns in words.

It may learn that words like:

- love
- excited
- great
- fun

often appear in happy messages.

It may also learn that words like:

- sad
- upset
- terrible
- angry

often appear in unhappy messages.

Then, when it sees a new sentence, it can make a guess.

That is the basic idea of AI.


## Part 13. What is machine learning?

Machine learning is one common way to build AI.

Machine learning means:

1. We collect examples
2. We give those examples to the computer
3. The computer looks for patterns
4. The computer uses those patterns to make predictions on new data

So if AI is the big idea, machine learning is one tool we use to make it happen.


## Part 14. What is a classifier?

In our first project, we will build a text classifier.

## Part 17. The library we will use

## Part 18. Build a simple happiness classifier

Read the comments carefully.

They explain what every important line is doing.


````python
# Import the classifier we want to use from the textblob library.
from textblob.classifiers import NaiveBayesClassifier

# Create a list of training examples.
# Each item in the list is a pair.
# The first part is the sentence.
# The second part is the label.
# 'pos' means the sentence sounds happy.
# 'neg' means the sentence sounds not happy.
training_data = [
    ("I am very happy today", "pos"),
    ("This is amazing", "pos"),
    ("I love this so much", "pos"),
    ("I feel great", "pos"),
    ("Today is awesome", "pos"),
    ("This is so much fun", "pos"),
    ("I am sad", "neg"),
    ("This is terrible", "neg"),
    ("I feel upset", "neg"),
    ("I am frustrated", "neg"),
    ("This is the worst", "neg"),
    ("I feel tired and angry", "neg")
]

# Create the classifier and train it using the examples above.
# At this moment, the model studies the examples and starts learning patterns.
classifier = NaiveBayesClassifier(training_data)

# Create a list of new sentences that the classifier has not seen before.
# These are the sentences we want the AI to predict.
test_sentences = [
    "I am so excited",
    "This is horrible",
    "I love learning Python",
    "I feel angry",
    "Today is a great day"
]

# Loop through each test sentence one at a time.
for sentence in test_sentences:
    # Ask the classifier to predict the label for this sentence.
    prediction = classifier.classify(sentence)

    # Print the sentence and the prediction.
    print(sentence, "->", prediction)
````

## Part 19. Make the prediction easier to read

Right now, the classifier prints `pos` or `neg`.

Let us translate those labels into words that are easier to read.


````python
# Use the same test sentences again.
for sentence in test_sentences:
    # Get the short label from the classifier.
    prediction = classifier.classify(sentence)

    # Convert the short label into a clearer message.
    if prediction == "pos":
        readable_label = "Happy"
    else:
        readable_label = "Not Happy"

    # Show the final result.
    print(sentence, "->", readable_label)
````

## Part 20. Let the student interact with the AI

Now it is your turn.

Type your own sentence and see what the classifier predicts.


````python
# Ask the user to type a sentence.
my_sentence = input("Type a sentence for the AI to classify: ")

# Get the prediction from the classifier.
my_prediction = classifier.classify(my_sentence)

# Convert the short label into a readable message.
if my_prediction == "pos":
    print("The AI thinks this sentence is: Happy")
else:
    print("The AI thinks this sentence is: Not Happy")
````
